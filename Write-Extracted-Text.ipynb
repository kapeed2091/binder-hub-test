```python
import csv
import os

FOLDER_PATH_CONTAINING_ORIGINAL_ID_CARDS = '/Users/kapeed2091/Desktop/Typeform-IIT-IDCards-06112019/Original'
FOLDER_PATH_CONTAINING_ROTATED_ID_CARDS = '/Users/kapeed2091/Desktop/Typeform-IIT-IDCards-06112019/Rotated'
FOLDER_PATH_CONTAINING_REKOGNITION_EXTRACTED_TEXT = '/Users/kapeed2091/Desktop/Typeform-IIT-IDCards-06112019/Rekognition'

FILE_TO_READ = '/Users/kapeed2091/Desktop/Typeform-Responses-With-Extracted-Data.csv'

import xlsxwriter 

workbook = xlsxwriter.Workbook('/Users/kapeed2091/Desktop/Typeform-Extracted-Text.xlsx') 
worksheet = workbook.add_worksheet() 


def get_text(file):
    print(file)
    if os.path.exists(file):
        f = open(file, 'r')
        return f.read().strip()
    else:
        return ''


with open(FILE_TO_READ, 'r') as csvfile:
    file_reader = csv.reader(csvfile, delimiter=',')
    index = 0
    for row in file_reader:
        index+=1
        
        if index == 1: #header row
            worksheet.write(index-1, 0, 'student_code')
            worksheet.write(index-1, 1, 'image')
            worksheet.write(index-1, 2, 'text1')
            worksheet.write(index-1, 3, 'text2')
            worksheet.write(index-1, 4, 'text3')
            worksheet.write(index-1, 5, 'text4')
            continue
            
        student_code = row[0].strip()
        image = row[6].strip()
        
        name, extension = os.path.splitext(image)
        
        file1 = os.path.join(FOLDER_PATH_CONTAINING_REKOGNITION_EXTRACTED_TEXT, '{}.txt'.format(name))
        file2 = os.path.join(FOLDER_PATH_CONTAINING_REKOGNITION_EXTRACTED_TEXT, '{}-{}.txt'.format(name, 90))
        file3 = os.path.join(FOLDER_PATH_CONTAINING_REKOGNITION_EXTRACTED_TEXT, '{}-{}.txt'.format(name, 180))
        file4 = os.path.join(FOLDER_PATH_CONTAINING_REKOGNITION_EXTRACTED_TEXT, '{}-{}.txt'.format(name, 270))
        
        text1 = get_text(file1)
        text2 = get_text(file2)
        text3 = get_text(file3)
        text4 = get_text(file4)
        
        worksheet.write(index-1, 0, student_code)
        worksheet.write(index-1, 1, image)
        worksheet.write(index-1, 2, text1)
        worksheet.write(index-1, 3, text2)
        worksheet.write(index-1, 4, text3)
        worksheet.write(index-1, 5, text4)

workbook.close()
```
