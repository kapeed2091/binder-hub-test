```python
FILE_TO_READ = '/Users/kapeed2091/Desktop/hotjar_metadata_04112019_1200-to-06112019_1130.csv'
MERGED_FILE_TO_WRITE = '/Users/kapeed2091/Desktop/merged-hotjar-stats.csv'

import csv
from collections import defaultdict

hotjar_user_dict = defaultdict(lambda: defaultdict(lambda: list()))

with open(FILE_TO_READ, 'r') as csvfile:
    file_reader = csv.reader(csvfile, delimiter=',')
    index = 0
    for row in file_reader:
        index+=1
        
        if index == 1: #header row
            continue
            
        student_code = row[0].strip()
        hotjar_video_id = row[1].strip()
        hotjar_user_id = row[2].strip()
        referrer = row[3].strip()
        landed_url = row[4].strip()
        exited_url = row[5].strip()
        num_pages = int(row[6])
        duration = int(row[7])
        date = row[8].strip()
        time = row[9].strip()
        recording_url = row[10].strip()
        
#         if hotjar_user_id not in hotjar_user_dict.keys():
#             hotjar_user_dict[hotjar_user_id] = de

        if student_code != '':
            hotjar_user_dict[hotjar_user_id]['student_code'].append(student_code)

        hotjar_user_dict[hotjar_user_id]['hotjar_video_id'].append(hotjar_video_id)
        hotjar_user_dict[hotjar_user_id]['referrer'].append(referrer)
        hotjar_user_dict[hotjar_user_id]['landed_url'].append(landed_url)
        hotjar_user_dict[hotjar_user_id]['exited_url'].append(exited_url)
        hotjar_user_dict[hotjar_user_id]['num_pages'].append(num_pages)
        hotjar_user_dict[hotjar_user_id]['duration'].append(duration)
        hotjar_user_dict[hotjar_user_id]['date'].append(date)
        hotjar_user_dict[hotjar_user_id]['time'].append(time)
        hotjar_user_dict[hotjar_user_id]['recording_url'].append(recording_url)

f=open(MERGED_FILE_TO_WRITE, 'w')
for hotjar_user_id in hotjar_user_dict.keys():
    print(hotjar_user_id)
    final_list = []
    user_dict = hotjar_user_dict[hotjar_user_id]
    final_list.append('|||'.join(list(set(user_dict['student_code']))))
    final_list.append('|||'.join(user_dict['hotjar_video_id']))
    final_list.append(hotjar_user_id)
    final_list.append('|||'.join(user_dict['referrer']))
    final_list.append('|||'.join(user_dict['landed_url']))
    final_list.append('|||'.join(user_dict['exited_url']))
    final_list.append(str(sum(user_dict['num_pages'])))
    final_list.append(str(sum(user_dict['duration'])))
    final_list.append('|||'.join(user_dict['date']))
    final_list.append('|||'.join(user_dict['time']))
    final_list.append('|||'.join(user_dict['recording_url']))
    
#     for ei in user_dict:
#         print('\t--->', ei)
#         print('\t\t--->--->', user_dict[ei])

    f.write(','.join(final_list)+'\n')
f.close()
```
